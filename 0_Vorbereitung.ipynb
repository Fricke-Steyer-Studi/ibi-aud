### GitLab

Das Lernmaterial in diesem Modulkurs wird Ihnen in *GitLab*
bereitgestellt, einem sogenannten *Source-Code-Management-System*
(SCM). Solche Systeme sind zum kollaborativen Arbeiten an
(typischerweise Software-)Projekten gedacht.  Programmierer:innen
können darin Quellcode abspeichern, Änderungen verfolgen und
rückgängig machen, Fehler melden und Code dokumentieren. GitLab ist
freie Software und wir nutzen die Instanz auf dem Server
http://scm.cms.hu-berlin.de/ die uns vom Computer- und Medienservice
(CMS) der Humboldt-Universität zu Berlin bereitgestellt wird.

Ein wesentlicher Baustein von GitLab ist *Git*, ein System zur
*verteilten Verwaltung von Dateiversionen*.

<div class="alert alert-success">

Bitte machen Sie sich mit GitLab vertraut, zum Beispiel mit Hilfe der
kurzen [Einführung zu
GitLab](https://pages.cms.hu-berlin.de/cms-webtech/gitlab-documentation/docs/home/)
vom CMS.

Library Carpentry bietet Ihnen eine [Einführung zu
Git](https://librarycarpentry.org/lc-git/), die allerdings über das
hinausgeht, was Sie zwingend für diesen Kurs benötigen. Sie können
diesen Kurs auch herunterladen, ohne Git zu installieren und **Sie
benötigen keinen GitHub-Account.**
    
</div>

Wenn Sie *Fehler* in den Kursmaterialien finden oder *Hinweise zur
Verbesserung* haben, können Sie diese in GitLab im Menü links unter
[Issues](https://scm.cms.hu-berlin.de/ibi/aud/issues) hinzufügen.

Sie können die Möglichkeiten des Systems später auch für Ihre *Haus-
oder Abschlussarbeiten* nutzen, indem Sie [ein eigenes Projekt

vertraut machen.

### Arbeiten mit diesem Kurs

Sie können sich die Notebooks direkt in GitLab in Ihrem Web-Browser
anschauen. Dabei fehlen jedoch Formatierungen und Bilder und vor allem
können Sie die interaktiven Beispiele nicht ausprobieren und einige
der Aufgaben nicht lösen. Daher müssen Sie sich die Notebooks auf
Ihren Rechner herunterladen und in Jupyter starten. Dazu gibt es im
Wesentlichen drei Möglichkeiten:

#### Notebooks einzeln herunterladen

Sie können jedes Notebook einzeln herunterladen, abspeichern und in
Jupyter öffnen. Dazu ist es wichtig, dass Sie das Notebook selbst
herunterladen und nicht die GitLab-Webseite.  Klicken Sie dazu auf der
Webseite mit dem jeweiligen Notebook auf den Button "Download" und
speichern Sie es als Datei:
![Download](https://amor.cms.hu-berlin.de/~jaeschkr/teaching/damostin/download_file.png)

Der Vorteil dieser Variante ist, dass Sie aktualisierte Notebooks
leicht einzeln herunterladen können und dass Sie keine weitere
Software benötigen.

#### Gesamtes Projekt herunterladen

Sie können [auf der
Projektstartseite](https://scm.cms.hu-berlin.de/ibi/aud) alle
Notebooks auf einmal [als ZIP-Datei
herunterladen](https://scm.cms.hu-berlin.de/ibi/aud/-/archive/master/aud-master.zip)
und auf Ihrem Rechner in ein Verzeichnis entpacken.

![Open raw](https://amor.cms.hu-berlin.de/~jaeschkr/teaching/damostin/download.png)

Der Vorteil dieser Variante ist, dass Sie alle Inhalte auf einmal in
einem Verzeichnis haben. Wenn Notebooks geändert oder hinzugefügt
werden, müssen Sie diesen Schritt wiederholen oder diese Notebooks
einzeln herunterladen.

#### Projekt per Git synchronisieren

Fortgeschrittene können sich dieses Projekt auch auf ihrem eigenen
Rechner mittels Git klonen und dann immer die aktuellste Version der
Notebooks herunterladen. Dazu müssen Sie Git installieren und sich
damit vertraut machen (die [Einführung zu Git von Library
Carpentry](https://librarycarpentry.org/lc-git/) kann Ihnen dabei
weiterhelfen).

<div class="alert alert-danger">

#### Hinweis

Im Verlauf des Kurses werden Sie die Notebooks verändern und
ergänzen. **Nutzen Sie die Notebooks als "lebendiges" Lernmaterial,

Achten Sie beim Aktualisieren der Notebooks darauf, dass Ihre

### Hilfe
Hilfe zu **Jupyter** finden Sie einerseits im Menüpunkt *Help* in
jedem Notebook. Andererseits können Sie die Taste `ESC` (Escape)
(links oben auf Ihrer Tastatur) und danach die Taste `h` drücken, um
eine Übersicht der gültigen Tastaturkürzel zu erhalten. Machen Sie
sich am besten mit einigen der Kürzel vertraut, das erleichtert die
spätere Arbeit ungemein. 

Folgende Kürzel im "Command Mode" sind besonders hilfreich:
- `Enter` wechselt in den Edit Mode
- `Strg-Enter` führt den Inhalt der Zelle aus
- `Y` versetzt die Zelle in den Python-Modus
- `M` versetzt die Zelle in den Markdown-Modus
- `A` fügt eine Zelle oberhalb ein
- `B` fügt eine Zelle unterhalb ein

Im "Edit Mode" sind folgende Kürzel besonders hilfreich:
- `ESC` wechselt in den Command Mode
- `Strg-Enter` führt den Inhalt der Zelle aus
- `Tab` (links neben der Taste `Q`) vervollständigt Code bzw. rückt
  Code ein
- `Shift-Tab` zeigt Informationen zum Python-Code an der Stelle des
  Cursors an

<div class="alert alert-success">

Anwendung von Programmbibliotheken zu vereinfachen, nachdem die

```

```

erhalten, haben Sie vermutlich das Notebook mit Ihrem Webbrowser als
HTML-Datei abgespeichert. Achten Sie darauf, es wie oben beschrieben
